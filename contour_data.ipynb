### Stap 4 — `bewoonde_niet_geïsoleerde_woning`

#### Uitleg

Stock bewoonde niet-geïsoleerde woningen per band.

#### Bron

`contour_vlaanderen_stocks.xlsx`, Brusselse inwoners per dB

#### Berekenmethode

1. Woningen Vlaanderen uit Excel.
2. Woningen Brussel = `inwoners_brussel / gem_inwoners_per_woning_vlaanderen` (fallback 2,0).
3. Totaal × `PCT_NIET_GEISOLEERD` (0,80).

```python
idx = lden.index
_inw_vla_lden = series_op_index(raw_vlaanderen_lden, 'inwoners', idx)
_inw_bru_lden = series_op_index(lden_bru_db, 'inwoners', idx)
_won_vla_lden = series_op_index(raw_vlaanderen_lden, 'aantal_woningen', idx)
_gem_inw_per_huis_lden = (_inw_vla_lden / _won_vla_lden.replace(0, pd.NA)).fillna(2.0)
_gem_inw_per_huis_lden = _gem_inw_per_huis_lden.where(_gem_inw_per_huis_lden > 0, 2.0)
_won_bru_lden = _inw_bru_lden / _gem_inw_per_huis_lden
won_totaal_lden = _won_vla_lden + _won_bru_lden
lden['bewoonde_niet_geïsoleerde_woning'] = won_totaal_lden * PCT_NIET_GEISOLEERD
_inw_vla_lnight = series_op_index(raw_vlaanderen_lnight, 'inwoners', idx)
_inw_bru_lnight = series_op_index(lnight_bru_db, 'inwoners', idx)
_won_vla_lnight = series_op_index(raw_vlaanderen_lnight, 'aantal_woningen', idx)
_gem_inw_per_huis_lnight = (_inw_vla_lnight / _won_vla_lnight.replace(0, pd.NA)).fillna(2.0)
_gem_inw_per_huis_lnight = _gem_inw_per_huis_lnight.where(_gem_inw_per_huis_lnight > 0, 2.0)
_won_bru_lnight = _inw_bru_lnight / _gem_inw_per_huis_lnight
won_totaal_lnight = _won_vla_lnight + _won_bru_lnight
lnight['bewoonde_niet_geïsoleerde_woning'] = won_totaal_lnight * PCT_NIET_GEISOLEERD
lden[["bewoonde_niet_geïsoleerde_woning"]].head()
```

## Vergunningstellers — jaarlijks gemiddelde (2020–2025)

Stappen **11–14** gebruiken dezelfde methode voor omgevingsloket, kwetsbare functies en verkaveling.

### Periode en eenheid

- **Geen enkel kalenderjaar** — we nemen het **jaarlijks gemiddelde** over **2020–2025** (6 jaren).
- Elke waarde in `vla` / `lden` is dus: *“gemiddeld aantal per jaar”*, geschikt als **flow-teller** (jaarlijkse instroom).

### Methode (per vergunningstype)

1. Filter gemeente-rijen in `2020–2025`; sluit `Totalen`-rijen en `handeling == "Totalen"` uit.
2. Per **type** (`bron`, `gemeente`, `handeling`, `gebouw_functie`, `metriek`): som per kalenderjaar.
3. **Gemiddelde** = som over 6 jaren ÷ 6 (ontbrekende jaren = 0).
4. `gemiddelde_jaarlijkse_vergunningen()` → label `jaar_indiening = "gemiddelde_2020_2025"`.
5. Daarna: gemeente → dB-contour via ruimtelijke conversietabel.

### Koppeling flow rates

| Variabele | Vergunningstype | Gebruikt in |
|---|---|---|
| `vergunde_wooneenheden_nieuwbouw` | Nieuwbouw × Aantal wooneenheden | `verbod_kleine_woning` |
| `vergunningen_kwetsbare_groep` | kwetsbare_functies | `verbod_kwetsbare_groep` |
| `renovatie_totaal` | Verbouwen of hergebruik | R+, R−, renovatiemaatregelen |
| `gem_wooneenheden_per_vergunning` | omgevingsloket (WE/projecten) | hulpcijfer; nog niet in flow-formules |

**Scope:** Vlaamse ringgemeenten die naar contour kunnen worden gemapt (niet heel Vlaanderen/Brussel).


### Stap 11 — `vergunde_wooneenheden_nieuwbouw`

#### Uitleg

**Jaarlijks gemiddelde** (2020–2025) van vergunde wooneenheden nieuwbouw per dB-band. Input voor flow rate `verbod_kleine_woning`.

#### Bron

`vergunningen_omgevingsloket_2026_lang.csv`

#### Berekenmethode

1. `verg_gem = gemiddelde_jaarlijkse_vergunningen(verg_combined)` — jaarlijks gemiddelde per type (2020–2025).
2. `verdeel_gemeente_naar_contour(verg_gem, …)` — verdeling naar dB-band.
3. Som per band over de rijen die bij deze variabele horen (filter op `handeling` / `metriek` / `bron`).

Filter: `handeling == 'Nieuwbouw'` en `metriek == 'Aantal wooneenheden'`.

```python
sector_lden, _ = lees_sector_contour_beide()
conversie_lden = conversietabel_gemeente_naar_db(sector_lden, indicator='lden')
conversie_lden_band = koppel_conversie_aan_contourband(conversie_lden, lden)
verg_omg, verg_kwets, verg_verk = lees_vergunningen()
verg_combined = combineer_vergunningen_lang(
    verg_omg.assign(bron='omgevingsloket'),
    verg_kwets.assign(bron='kwetsbare_functies'),
    verg_verk.assign(bron='verkaveling'),
)
verg_gem = gemiddelde_jaarlijkse_vergunningen(verg_combined)
verg_contour, _verg_niet = verdeel_gemeente_naar_contour(
    verg_gem, conversie_lden_band, jaartal=JAAR_TELLERS
)
idx = lden.index
df_verg = verg_contour[verg_contour['jaar_indiening'].astype(str) == str(JAAR_TELLERS)].copy()

def som_per_band(mask):
    subset = df_verg.loc[mask]
    if subset.empty:
        return pd.Series(0.0, index=idx)
    return subset.groupby('db_ondergrens')['waarde'].sum().reindex(idx).fillna(0)

_nieuwbouw_we = som_per_band(
    (df_verg['handeling'] == 'Nieuwbouw') & (df_verg['metriek'] == 'Aantal wooneenheden')
)
lden['vergunde_wooneenheden_nieuwbouw'] = _nieuwbouw_we
lnight['vergunde_wooneenheden_nieuwbouw'] = _nieuwbouw_we
lden[["vergunde_wooneenheden_nieuwbouw"]].head()
```

### Stap 15 — `transacties_en_capakey_prijzen`

1. Prijs per transactie (`prijs_euro`).
2. Gewogen gemiddelde prijs per CaPaKey per prijstype.
3. CaPaKey → contour via extern bestand + Brusselse sectorfallback.

```python
from contour.prices import (
    SEGMENTEN_PER_PRIJSKOLOM,
    bouw_capakey_contour_mapping,
)

# 1) Prijs per transactie
tx = transacties.copy()
tx['capakey'] = tx['capakey'].astype(str)
tx['aantal_transacties'] = pd.to_numeric(tx['sum_ParcelsNumber'], errors='coerce').fillna(0)
p50 = pd.to_numeric(tx['avg_PriceP50'], errors='coerce')
m2 = pd.to_numeric(tx['average_price_m2'], errors='coerce')
opp = pd.to_numeric(tx['avg_ParcelsAreaP50'], errors='coerce')
tx['prijs_euro'] = p50.where(p50 > 0, m2 * opp)
tx['prijs_euro'] = tx['prijs_euro'].where(tx['prijs_euro'] > 0)

# 2) Gewogen gemiddelde prijs per CaPaKey (per prijstype)
capakey_prijzen = pd.DataFrame({'capakey': tx['capakey'].unique()})
for _prijs_kolom, _segmenten in SEGMENTEN_PER_PRIJSKOLOM.items():
    seg = tx[tx['segment'].isin(_segmenten)]
    _gewicht_kolom = f'gewicht_{_prijs_kolom}'
    totaal_tx = seg.groupby('capakey', as_index=False)['aantal_transacties'].sum()
    totaal_tx = totaal_tx.rename(columns={'aantal_transacties': _gewicht_kolom})
    met_prijs = seg[seg['prijs_euro'].notna() & (seg['aantal_transacties'] > 0)].copy()
    if met_prijs.empty:
        met_prijs = seg[seg['prijs_euro'].notna()].copy()
    if met_prijs.empty:
        capakey_prijzen = capakey_prijzen.merge(
            totaal_tx.assign(**{_prijs_kolom: pd.NA}), on='capakey', how='left'
        )
        continue
    met_prijs['_w'] = met_prijs['aantal_transacties'].clip(lower=1)
    prijs_agg = (
        met_prijs.assign(_pw=met_prijs['prijs_euro'] * met_prijs['_w'])
        .groupby('capakey', as_index=False)
        .agg(_pw=('_pw', 'sum'), _w=('_w', 'sum'))
    )
    prijs_agg[_prijs_kolom] = prijs_agg['_pw'] / prijs_agg['_w']
    capakey_prijzen = capakey_prijzen.merge(
        totaal_tx.merge(prijs_agg[['capakey', _prijs_kolom]], on='capakey', how='left'),
        on='capakey',
        how='left',
    )

# Geen iso-split: geïsoleerde woning = zelfde prijs
capakey_prijzen['prijs_bewoonde_geïsoleerde_woning'] = capakey_prijzen[
    'prijs_bewoonde_niet_geïsoleerde_woning'
]
capakey_prijzen['gewicht_prijs_bewoonde_geïsoleerde_woning'] = capakey_prijzen[
    'gewicht_prijs_bewoonde_niet_geïsoleerde_woning'
]

# 3) CaPaKey → db_ondergrens
capakey_mapping = bouw_capakey_contour_mapping(transacties, brussel_lden, lden)

def agg_transacties(segmenten):
    gekoppeld = tx.merge(capakey_mapping[['capakey', 'db_ondergrens']], on='capakey', how='inner')
    subset = gekoppeld[gekoppeld['segment'].isin(segmenten)]
    if subset.empty:
        return pd.Series(0.0, index=lden.index)
    return subset.groupby('db_ondergrens')['aantal_transacties'].sum().reindex(lden.index).fillna(0)

capakey_prijzen.head()
```

### Stap 20 — `prijs_onbebouwde_bebouwbare_percelen`

#### Uitleg

FLOW-variabele `prijs_onbebouwde_bebouwbare_percelen`.

#### Bron

Zie pipeline-bronnen in `contour/`

#### Berekenmethode

Zie STOCKS_EN_FLOWS_BEREKENEN.md §2.

```python
_prijs_kolom = 'prijs_onbebouwde_bebouwbare_percelen'
_gewicht_tx = f'gewicht_{_prijs_kolom}'
_gekoppeld = capakey_mapping.merge(
    capakey_prijzen[['capakey', _prijs_kolom, _gewicht_tx]],
    on='capakey',
    how='inner',
)
_gekoppeld = _gekoppeld[_gekoppeld[_prijs_kolom].notna()].copy()
_gekoppeld['w'] = _gekoppeld['gewicht_ruimtelijk'] * _gekoppeld[_gewicht_tx].clip(lower=1)

def _gewogen_gem(prijzen, gewichten):
    mask = prijzen.notna() & (gewichten > 0)
    if not mask.any():
        return None
    return float((prijzen[mask] * gewichten[mask]).sum() / gewichten[mask].sum())

_fallback = _gewogen_gem(
    capakey_prijzen[_prijs_kolom],
    capakey_prijzen[_gewicht_tx].clip(lower=1),
)
_per_band = _gekoppeld.groupby('db_ondergrens').apply(
    lambda g: _gewogen_gem(g[_prijs_kolom], g['w']),
    include_groups=False,
)
lden[_prijs_kolom] = _per_band.reindex(lden.index)
if _fallback is not None:
    lden[_prijs_kolom] = lden[_prijs_kolom].fillna(_fallback)
lnight[_prijs_kolom] = lden[_prijs_kolom]
lden[['prijs_onbebouwde_bebouwbare_percelen']].head()
```

### Stap 22 — `prijs_bewoonde_niet_geïsoleerde_woning`

#### Uitleg

Eenheidsprijs bewoonde niet-geïsoleerde woning per band (€).

#### Bron

`transacties_woningen.csv`, `transacties_appartementen.csv`, CaPaKey-contour mapping

#### Berekenmethode

1. Transactieprijs per CaPaKey: mediaan `avg_PriceP50`, anders `average_price_m2 × avg_ParcelsAreaP50`.
2. Segmenten `woningen` + `appartementen`: gewogen gemiddelde prijs per CaPaKey (gewicht = aantal transacties).
3. Koppel CaPaKey → `db_ondergrens` via `capakey_mapping` (ruimtelijk gewicht × transactiegewicht).
4. Gewogen gemiddelde prijs per band; fallback = landelijk gewogen gemiddelde.

```python
_prijs_kolom = 'prijs_bewoonde_niet_geïsoleerde_woning'
_gewicht_tx = f'gewicht_{_prijs_kolom}'
_gekoppeld = capakey_mapping.merge(
    capakey_prijzen[['capakey', _prijs_kolom, _gewicht_tx]],
    on='capakey',
    how='inner',
)
_gekoppeld = _gekoppeld[_gekoppeld[_prijs_kolom].notna()].copy()
_gekoppeld['w'] = _gekoppeld['gewicht_ruimtelijk'] * _gekoppeld[_gewicht_tx].clip(lower=1)

def _gewogen_gem(prijzen, gewichten):
    mask = prijzen.notna() & (gewichten > 0)
    if not mask.any():
        return None
    return float((prijzen[mask] * gewichten[mask]).sum() / gewichten[mask].sum())

_fallback = _gewogen_gem(
    capakey_prijzen[_prijs_kolom],
    capakey_prijzen[_gewicht_tx].clip(lower=1),
)
_per_band = _gekoppeld.groupby('db_ondergrens').apply(
    lambda g: _gewogen_gem(g[_prijs_kolom], g['w']),
    include_groups=False,
)
lden[_prijs_kolom] = _per_band.reindex(lden.index)
if _fallback is not None:
    lden[_prijs_kolom] = lden[_prijs_kolom].fillna(_fallback)
lnight[_prijs_kolom] = lden[_prijs_kolom]
lden[['prijs_bewoonde_niet_geïsoleerde_woning']].head()
```
