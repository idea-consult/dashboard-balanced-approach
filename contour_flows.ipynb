# Contour flows — flow rates per dB-band

Leest parquet uit `output/intermediate/` (gegenereerd door `contour_data.ipynb`).

Elke maatregel krijgt **twee kolommen** in `lden_flows`: `{measure}_baseline` en `{measure}_active`, berekend **per `db_ondergrens`** uit kolommen van `lden_vars`.

Referentie: [STOCKS_EN_FLOWS_BEREKENEN.md §4](../STOCKS_EN_FLOWS_BEREKENEN.md) + `contour/flows_per_contour.py`.

```python
from pathlib import Path
import pandas as pd

INTERMEDIATE = Path('output/intermediate')
contour_lden = pd.read_parquet(INTERMEDIATE / 'lden.parquet')
vergunningen_contour = pd.read_parquet(INTERMEDIATE / 'vergunningen_contour.parquet')
transacties_capakey = pd.read_parquet(INTERMEDIATE / 'transacties_capakey.parquet')
```

## Deel A — Variabelen per band (`lden_vars`)

Alle FLOW-kolommen uit `lden.parquet`, index = `db_ondergrens`.

```python
from contour.flows_per_contour import lden_vars_bereiden

lden_vars = lden_vars_bereiden(contour_lden)
lden_flows = pd.DataFrame(index=lden_vars.index)
lden_vars.head()
```

## Deel B — Flow rates stap voor stap (STOCKS_EN_FLOWS §4)

Elke stap schrijft expliciet `{measure}_baseline` en `{measure}_active` uit kolommen van `lden_vars` (teller / noemer, veilig bij noemer 0).

## Verkavelingsverbod

### Uitleg

Verbod op verkaveling in de geluidszone. In het simulatiemodel heeft deze maatregel **geen stock-transfer**: geen woningen of percelen veranderen van categorie.

### Stock-flow

`—`

### Berekening

- **Baseline:** `0 (geen effect)`
- **Active:** `0 (geen effect)`

### Aannames

Geen quantitatief hinder-effect gemodelleerd — maatregel staat wel in `flow_rules.csv` voor kostentoewijzing (indien ingevuld).

### Gebruikte variabelen

- Geen (vaste waarde of geen effect)

```python
lden_flows["verkavelingsverbod_baseline"] = 0.0
lden_flows["verkavelingsverbod_active"] = 0.0
lden_flows[["verkavelingsverbod_baseline", "verkavelingsverbod_active"]].head(10)
```

## Woongebiedverbod

### Uitleg

Schrappen van woongebied-aanduidingen: percelen gaan van **onbebouwd bebouwbare** naar **onbebouwd onbebouwbare** stock (omgekeerd van woongebied-aanduiding). Baseline = historische netto-aanduiding; active = versnelde schrapping.

### Stock-flow

`onbebouwde_onbebouwbare_percelen ↔ onbebouwde_bebouwbare_percelen`

### Berekening

- **Baseline:** `(bebouwbare_percelen_woongebied(5jr) − niet_bebouwbare_percelen_woongebied_schrapping(5jr)) / 5 / onbebouwde_onbebouwbare_percelen`
- **Active:** `niet_bebouwbare_percelen_woongebied_schrapping(5jr) / 5 / onbebouwde_onbebouwbare_percelen`

### Aannames

Cumulatieve 5-jaar-tellers worden **geannualiseerd** (/ 5).
Woongebied-data alleen uit Vlaanderen-contour; Brussel = 0.
`onbebouwde_onbebouwbare_percelen` is nu **placeholder 0** → rate = 0 tot echte stock per band.

### Gebruikte variabelen

- `bebouwbare_percelen_woongebied(5jr)`
- `niet_bebouwbare_percelen_woongebied_schrapping(5jr)`
- `onbebouwde_onbebouwbare_percelen`

```python
lden_flows["woongebiedverbod_baseline"] = (
    (lden_vars["bebouwbare_percelen_woongebied(5jr)"] - lden_vars["niet_bebouwbare_percelen_woongebied_schrapping(5jr)"]) / 5
).div(lden_vars["onbebouwde_onbebouwbare_percelen"]).where(lden_vars["onbebouwde_onbebouwbare_percelen"] > 0, 0).clip(upper=1)
lden_flows["woongebiedverbod_active"] = (
    lden_vars["niet_bebouwbare_percelen_woongebied_schrapping(5jr)"] / 5
).div(lden_vars["onbebouwde_onbebouwbare_percelen"]).where(lden_vars["onbebouwde_onbebouwbare_percelen"] > 0, 0).clip(upper=1)
lden_flows[["woongebiedverbod_baseline", "woongebiedverbod_active"]].head(10)
```

## Aankoopbeleid percelen

### Uitleg

Overheid koopt onbebouwde bebouwbare percelen op. In actief scenario stijgt `perceel_eigendom_overheid`; de privé-stock `onbebouwde_bebouwbare_percelen` daalt.

### Stock-flow

`onbebouwde_bebouwbare_percelen → perceel_eigendom_overheid`

### Berekening

- **Baseline:** `0 (geen aankoopbeleid in referentie)`
- **Active:** `0.25 (= 50% publiek × 50% opkoopbaar) × alle_transacties_percelen / onbebouwde_bebouwbare_percelen`

### Aannames

Opkoopaandeel = **0.25** (50% publieke actor × 50% van transacties die opkoopbaar zijn).
Transactieteller = industrie/onbebouwd segment (`industrie_terrein` + `industrie_bebouwd`), gekoppeld per CaPaKey → contour.
Geen aparte filter op «bebouwbaar» in transactiedata.
`onbebouwde_bebouwbare_percelen` placeholder **0** → rate = 0.

### Gebruikte variabelen

- `alle_transacties_percelen`
- `onbebouwde_bebouwbare_percelen`

```python
_noemer = lden_vars["onbebouwde_bebouwbare_percelen"]
_teller = 0.25 * lden_vars["alle_transacties_percelen"]
lden_flows["aankoopbeleid_percelen_baseline"] = 0.0
lden_flows["aankoopbeleid_percelen_active"] = (_teller).div(_noemer).where(_noemer > 0, 0).clip(upper=1)
lden_flows[["aankoopbeleid_percelen_baseline", "aankoopbeleid_percelen_active"]].head(10)
```

## Voorkooprecht percelen

### Uitleg

Gemeente/overheid oefent voorkooprecht uit bij verkoop van onbebouwde bebouwbare percelen. Teller = alle verkopen (≈ transacties) per band.

### Stock-flow

`onbebouwde_bebouwbare_percelen → perceel_eigendom_overheid`

### Berekening

- **Baseline:** `0`
- **Active:** `alle_verkopen_onbebouwde_bebouwbare_percelen / onbebouwde_bebouwbare_percelen`

### Aannames

**Geen publiek/privé-split** in transactiedata → `alle_verkopen_onbebouwde_bebouwbare_percelen` = `alle_transacties_percelen` (zelfde waarde).
Voorkooprecht geldt als 100% van die transactiestroom in actief scenario.
Stock-noemer placeholder 0.

### Gebruikte variabelen

- `alle_verkopen_onbebouwde_bebouwbare_percelen`
- `onbebouwde_bebouwbare_percelen`

```python
_noemer = lden_vars["onbebouwde_bebouwbare_percelen"]
_teller = lden_vars["alle_verkopen_onbebouwde_bebouwbare_percelen"]
lden_flows["voorkooprecht_percelen_baseline"] = 0.0
lden_flows["voorkooprecht_percelen_active"] = (_teller).div(_noemer).where(_noemer > 0, 0).clip(upper=1)
lden_flows[["voorkooprecht_percelen_baseline", "voorkooprecht_percelen_active"]].head(10)
```

## Verbod kleine woning

### Uitleg

Verbod op kleine/MER-vrije nieuwbouwwoningen in de zone. Baseline én active gebruiken dezelfde rate: vergunde wooneenheden nieuwbouw t.o.v. onbebouwde bebouwbare percelen (proportioneel per band).

### Stock-flow

`onbebouwde_bebouwbare_percelen → nieuwe_woning (via simulator)`

### Berekening

- **Baseline:** `vergunde_wooneenheden_nieuwbouw / onbebouwde_bebouwbare_percelen`
- **Active:** `= baseline (zelfde formule)`

### Aannames

Vergunningen per gemeente → contour via conversietabel (deels Vlaamse ring).
Filter: `Nieuwbouw` + `Aantal wooneenheden`, jaar = 2025.
**Geen MER-split** in brondata: totaal nieuwbouw-wooneenheden als proxy voor «kleine» woning.

### Gebruikte variabelen

- `vergunde_wooneenheden_nieuwbouw`
- `onbebouwde_bebouwbare_percelen`

```python
_noemer = lden_vars["onbebouwde_bebouwbare_percelen"]
_teller = lden_vars["vergunde_wooneenheden_nieuwbouw"]
_rate = (_teller).div(_noemer).where(_noemer > 0, 0).clip(upper=1)
lden_flows["verbod_kleine_woning_baseline"] = _rate
lden_flows["verbod_kleine_woning_active"] = _rate
lden_flows[["verbod_kleine_woning_baseline", "verbod_kleine_woning_active"]].head(10)
```

## Renovatie zonder maatregel

### Uitleg

Referentie-renovatiestroom **met** akoestische/isolatiemaatregel (R+): woningen gaan van niet-geïsoleerd naar geïsoleerd **zonder** extra beleidsimpuls.

### Stock-flow

`bewoonde_niet_geïsoleerde_woning → bewoonde_geïsoleerde_woning`

### Berekening

- **Baseline:** `R+ / bewoonde_niet_geïsoleerde_woning  waar R+ = renovatie_totaal × 0.5 (R+) en × 0.5 (R−)`
- **Active:** `0`

### Aannames

Geen iso-split in vergunningen → **0.5** van `renovatie_totaal` = R+, rest = R−.
Baseline behoudt «natuurlijke» R+-stroom; active = 0 (geen extra maatregel).

### Gebruikte variabelen

- `renovatie_totaal`
- `bewoonde_niet_geïsoleerde_woning`

```python
_renovatie = lden_vars["renovatie_totaal"]
_r_plus = _renovatie * 0.5
_noemer = lden_vars["bewoonde_niet_geïsoleerde_woning"]
lden_flows["renovatie_zonder_maatregel_baseline"] = (_r_plus).div(_noemer).where(_noemer > 0, 0).clip(upper=1)
lden_flows["renovatie_zonder_maatregel_active"] = 0.0
lden_flows[["renovatie_zonder_maatregel_baseline", "renovatie_zonder_maatregel_active"]].head(10)
```

## Verplicht isoleren bij renovatie

### Uitleg

Verplichte isolatie bij elke renovatie: alle renovaties (R+ + R−) moeten isoleren in actief scenario.

### Stock-flow

`bewoonde_niet_geïsoleerde_woning → bewoonde_geïsoleerde_woning`

### Berekening

- **Baseline:** `0`
- **Active:** `(R+ + R−) / bewoonde_niet_geïsoleerde_woning`

### Aannames

R+ + R− = volledige `renovatie_totaal` (vergunningen «Verbouwen of hergebruik»).
Split R+/R− via **0.5** alleen voor baseline van andere maatregelen.

### Gebruikte variabelen

- `renovatie_totaal`
- `bewoonde_niet_geïsoleerde_woning`

```python
_renovatie = lden_vars["renovatie_totaal"]
_r_plus = _renovatie * 0.5
_r_min = _renovatie * 0.5
_noemer = lden_vars["bewoonde_niet_geïsoleerde_woning"]
lden_flows["verplicht_isoleren_renovatie_baseline"] = 0.0
lden_flows["verplicht_isoleren_renovatie_active"] = (_r_plus + _r_min).div(_noemer).where(_noemer > 0, 0).clip(upper=1)
lden_flows[["verplicht_isoleren_renovatie_baseline", "verplicht_isoleren_renovatie_active"]].head(10)
```

## Gesubsidieerd isolatieprogramma

### Uitleg

Subsidie stimuleert isolatie: **dubbele** renovatiestroom t.o.v. verplicht isoleren.

### Stock-flow

`bewoonde_niet_geïsoleerde_woning → bewoonde_geïsoleerde_woning`

### Berekening

- **Baseline:** `0`
- **Active:** `2 × (R+ + R−) / bewoonde_niet_geïsoleerde_woning`

### Aannames

Factor **2×** op totale renovatiestroom (beleidsintensiteit gesubsidieerd programma).

### Gebruikte variabelen

- `renovatie_totaal`
- `bewoonde_niet_geïsoleerde_woning`

```python
_renovatie = lden_vars["renovatie_totaal"]
_r_plus = _renovatie * 0.5
_r_min = _renovatie * 0.5
_noemer = lden_vars["bewoonde_niet_geïsoleerde_woning"]
lden_flows["gesubsidieerd_isolatieprogramma_baseline"] = 0.0
lden_flows["gesubsidieerd_isolatieprogramma_active"] = (2.0 * (_r_plus + _r_min)).div(_noemer).where(_noemer > 0, 0).clip(upper=1)
lden_flows[["gesubsidieerd_isolatieprogramma_baseline", "gesubsidieerd_isolatieprogramma_active"]].head(10)
```

## Gestuurd isolatieprogramma

### Uitleg

Actief, gericht isolatieprogramma: **viervoudige** renovatiestroom.

### Stock-flow

`bewoonde_niet_geïsoleerde_woning → bewoonde_geïsoleerde_woning`

### Berekening

- **Baseline:** `0`
- **Active:** `4 × (R+ + R−) / bewoonde_niet_geïsoleerde_woning`

### Aannames

Factor **4×** — sterkste isolatie-intensiteit in het maatregelenpakket.

### Gebruikte variabelen

- `renovatie_totaal`
- `bewoonde_niet_geïsoleerde_woning`

```python
_renovatie = lden_vars["renovatie_totaal"]
_r_plus = _renovatie * 0.5
_r_min = _renovatie * 0.5
_noemer = lden_vars["bewoonde_niet_geïsoleerde_woning"]
lden_flows["gestuurd_isolatieprogramma_baseline"] = 0.0
lden_flows["gestuurd_isolatieprogramma_active"] = (4.0 * (_r_plus + _r_min)).div(_noemer).where(_noemer > 0, 0).clip(upper=1)
lden_flows[["gestuurd_isolatieprogramma_baseline", "gestuurd_isolatieprogramma_active"]].head(10)
```

## Aanleg geluidsbuffers

### Uitleg

Akoestische isolatie via geluidsbuffers: woningen worden geïsoleerd zonder klassieke renovatie-flow.

### Stock-flow

`bewoonde_niet_geïsoleerde_woning → bewoonde_geïsoleerde_woning`

### Berekening

- **Baseline:** `0`
- **Active:** `potentieel_isoleerbare_woningen / 5 / bewoonde_niet_geïsoleerde_woning`

### Aannames

`potentieel_isoleerbare_woningen` = **placeholder 0** (nog niet gemodelleerd) → rate = 0.
Annualisatie / 5 analoge aan 5-jaar-tellers.

### Gebruikte variabelen

- `potentieel_isoleerbare_woningen`
- `bewoonde_niet_geïsoleerde_woning`

```python
_noemer = lden_vars["bewoonde_niet_geïsoleerde_woning"]
_teller = lden_vars["potentieel_isoleerbare_woningen"] / 5
lden_flows["aanleg_geluidsbuffers_baseline"] = 0.0
lden_flows["aanleg_geluidsbuffers_active"] = (_teller).div(_noemer).where(_noemer > 0, 0).clip(upper=1)
lden_flows[["aanleg_geluidsbuffers_baseline", "aanleg_geluidsbuffers_active"]].head(10)
```
