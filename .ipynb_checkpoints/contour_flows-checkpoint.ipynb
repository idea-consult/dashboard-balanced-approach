# Contour flows — stocks, tellers en flow rates

Leest parquet uit `output/intermediate/` (gegenereerd door `contour_data.ipynb`).

Referentie: [FLOW_BEREKENINGEN.md §3](../FLOW_BEREKENINGEN.md) + `input/flow_rules.csv`.

```python
from pathlib import Path
import pandas as pd

INTERMEDIATE = Path('output/intermediate')
contour_lden = pd.read_parquet(INTERMEDIATE / 'contour_lden.parquet')
vergunningen_contour = pd.read_parquet(INTERMEDIATE / 'vergunningen_contour.parquet')
transacties_capakey = pd.read_parquet(INTERMEDIATE / 'transacties_capakey.parquet')
```

## Deel D — Stocks (FLOW §2.1)

```python
from contour.flows import bouw_stocks_contour

stocks_contour = bouw_stocks_contour(contour_lden)
stocks_contour.head()
```

Placeholder-vastgoedstock (D.3–D.7): 20% geo / 80% niet-geo woningen; te vervangen wanneer echte vastgoeddata per contour beschikbaar is.

## Deel E — Tellers (FLOW §2.2–2.4)

```python
from contour.flows import bouw_tellers_contour

tellers_contour = bouw_tellers_contour(contour_lden, vergunningen_contour, transacties_capakey)
tellers_contour.query('contour == "_totaal"')
```

## Deel F — Flow rates (FLOW §3)

```python
from contour.flows import bereken_alle_flow_rates, valideer_flow_rates

flow_rates = bereken_alle_flow_rates(contour_lden, vergunningen_contour, transacties_capakey, tellers_contour)
validatie = valideer_flow_rates(flow_rates)
flow_rates
```
